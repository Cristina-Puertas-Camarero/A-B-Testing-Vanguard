## Evaluación de duración

¿Fue el período de tiempo del experimento (del 15/3/2017 al 20/6/2017) adecuado para recopilar datos y conocimientos significativos?

```python
# Convertir la columna date_time a formato datetime
df['date_time'] = pd.to_datetime(df['date_time'])

# Filtrar las fechas del experimento
start_date = '2017-03-15'
end_date = '2017-06-20'
df_experiment = df[(df['date_time'] >= start_date) & (df['date_time'] <= end_date)]

# Contar el número de visitas por día
visits_per_day = df_experiment['date_time'].dt.date.value_counts().sort_index()

# Graficar la cantidad de visitas por día
plt.figure(figsize=(12, 6))
visits_per_day.plot(kind='bar')
plt.xlabel('Fecha')
plt.ylabel('Número de visitas')
plt.title('Número de visitas por día durante el experimento')
plt.show()
```

Si, ya que observamos, ua concentracion de visitas en torno a un mes, abril y mayo.
